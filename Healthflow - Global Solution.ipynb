**Objetivo do Modelo**  
O propósito do modelo a ser construído, tendo em vista o conjunto de dados disponibilizado, é efetuar uma análise da classificação de doenças com base nos sintomas apresentados pelo usuário. A finalidade dessa análise é discernir grupos distintos de usuários que compartilhem sintomas semelhantes, a fim de fornecer suporte em conjunto ao atendente disponível para avaliar a urgência do atendimento. **Tamanho**

**Quantidade de Registros**  
4920 registros de sintomas associados à doenças.

**Linhas x Colunas**  
4920 linhas (registros) e 18 colunas (atributos). As colunas são de Doença e até 17 sintomas associados.

**Dados Faltantes**

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Carregar dados do arquivo CSV
df = pd.read_csv("dataset_pt.csv")

# Contagem de ocorrências de cada sintoma
symptom_counts = df.iloc[:, 1:].stack().value_counts()

# Selecionar os 10 primeiros sintomas
top_10_symptoms = symptom_counts.head(10)

# Gráfico de barras horizontais
plt.figure(figsize=(10, 6))
sns.barplot(x=top_10_symptoms.values, y=top_10_symptoms.index, palette='pastel')

# Ajuste de tamanho de fonte para o eixo y
plt.yticks(fontsize=12)  # Ajuste de tamanho de fonte

plt.title('Contagem de Ocorrências dos 10 Sintomas Mais Frequentes', fontsize=16)
plt.xlabel('Contagem', fontsize=14)
plt.ylabel('Sintomas', fontsize=14)

# Adicionar os valores acima das barras
for index, value in enumerate(top_10_symptoms.values):
    plt.text(value, index, str(value), ha='left', va='center', fontsize=10)

plt.show()
```

```python
from sklearn.model_selection import train_test_split

data = df.iloc[:,1:].values
labels = df['Doença'].values

x_train, x_test, y_train, y_test = train_test_split(data, labels, train_size = 0.8,random_state=42)
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

**Primeiro Modelo - Decision Tree**  

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score, accuracy_score, confusion_matrix,classification_report, ConfusionMatrixDisplay,RocCurveDisplay ,precision_score,roc_curve
from sklearn.model_selection import train_test_split,KFold,cross_val_score,GridSearchCV
import matplotlib.pyplot as plt


tree =DecisionTreeClassifier(criterion='gini',random_state=42,max_depth=13)
tree.fit(x_train, y_train)
preds=tree.predict(x_test)
conf_mat = confusion_matrix(y_test, preds)
df_cm = pd.DataFrame(conf_mat, index=df['Doença'].unique(), columns=df['Doença'].unique())
print('F1-score% =', f1_score(y_test, preds, average='macro')*100, '|', 'Accuracy% =', accuracy_score(y_test, preds)*100)
sns.heatmap(df_cm)
```
